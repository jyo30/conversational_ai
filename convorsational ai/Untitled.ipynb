```python
import openai
import time

openai.api_key = None  # Set to None to use the free GPT-3 Playground

def ask_gpt3(prompt):
    completions = openai.Completion.create(
        engine="text-davinci-002",
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )
    message = completions.choices[0].text.strip()
    return message

print("AI: Hello! How can I help you today?")
while True:
    prompt = input("Human: ")
    start = time.time()
    response = ask_gpt3(prompt)
    end = time.time()
    print(f"AI: {response} (in {end - start:.2f} seconds)")
```

```python
import random

def get_response(input_text):
    """
    This function takes in a string of input text from the user, and generates a response
    using a list of pre-defined responses.
    """
    input_text = input_text.lower()
    greetings = ["hi", "hello", "hey", "what's up"]
    responses = ["Interesting!", "Tell me more.", "I'm not sure I understand.", "That's a good point."]
    
    if input_text in greetings:
        return random.choice(["Hello!", "Hi there!", "Hey!"])
    else:
        return random.choice(responses)

print("AI: Hello! How can I help you?")
while True:
    user_input = input("Human: ")
    if user_input.lower() in ["bye", "goodbye"]:
        print("AI: Goodbye!")
        break
    else:
        response = get_response(user_input)
        print("AI: " + response)
```
